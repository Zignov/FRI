### Evaluating Object Detection Algorithms: Precision

***Precision = True Positives (TP) / (False Positives (FP) + True Positives (TP)​)***

```python
# Simulated ground truth and predicted bounding box data
# Each bounding box is represented as [x1, y1, x2, y2]

# Ground truth bounding boxes represented in thick green color
ground_truth_boxes = [[113, 217, 289, 410], [441, 212, 609, 401]]
# Predicted bounding boxes represented in thin blue color
predicted_boxes = [[113, 217, 289, 410], [328, 79, 505, 266], [364, 302, 477, 432]]

def calculate_precision(gt_boxes, pred_boxes):
    tp = 0
    fp = 0
    
    for box in pred_boxes:
        if box in gt_boxes:
            tp += 1
        else:
            fp += 1
    
    precision = tp / (tp + fp)
    return precision

# Load an image
image_path = "data/fruit1.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (700,500))

# Create a blank image using matplotlib
fig, ax = plt.subplots(1, figsize=(8,8))
ax.set_title(f"Precision: {calculate_precision(ground_truth_boxes, predicted_boxes):.2f}")
plt.imshow(image)

# Draw ground truth boxes in green
for box in ground_truth_boxes:
    rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=6, edgecolor='g', facecolor='none', label='Ground Truth')
    ax.add_patch(rect)

# Draw predicted boxes in blue
for box in predicted_boxes:
    rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=1, edgecolor='b', facecolor='none', label='Prediction')
    ax.add_patch(rect)

# Remove axis for better visualization
ax.axis('off')

# Show the image
plt.show()
```

### Evaluating Object Detection Algorithms: Recall

***Recall = True Positives (TP) / (False Negatives (FN) + True Positives (TP)​)***

```python
# Simulated ground truth and predicted bounding box data
# Each bounding box is represented as [x1, y1, x2, y2]

ground_truth_boxes = [[113, 217, 289, 410], [441, 212, 609, 401]]
predicted_boxes = [[113, 217, 289, 410], [328, 79, 505, 266], [364, 302, 477, 432]]

def calculate_recall(gt_boxes, pred_boxes):
    tp = 0
    fn = 0
    
    for box in gt_boxes:
        if box in pred_boxes:
            tp += 1
        else:
            fn += 1
    
    recall = tp / (tp + fn)
    return recall

# Load an image
image_path = "data/fruit1.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (700,500))

# Create a blank image using matplotlib
fig, ax = plt.subplots(1, figsize=(8,8))
ax.set_title(f"Recall: {calculate_recall(ground_truth_boxes, predicted_boxes):.2f}")
plt.imshow(image)

# Draw ground truth boxes in green
for box in ground_truth_boxes:
    rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=6, edgecolor='g', facecolor='none', label='Ground Truth')
    ax.add_patch(rect)

# Draw predicted boxes in blue
for box in predicted_boxes:
    rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=1, edgecolor='b', facecolor='none', label='Prediction')
    ax.add_patch(rect)

# Remove axis for better visualization
ax.axis('off')

# Show the image
plt.show()
```

### Evaluating Object Detection Algorithms: F1 Score

***F1 Score = 2 × (Precision + Recall) / (Precision × Recall​)***

```python
# Simulated ground truth and predicted bounding box data
# Each bounding box is represented as [x1, y1, x2, y2]

ground_truth_boxes = [[113, 217, 289, 410], [441, 212, 609, 401]]
predicted_boxes = [[113, 217, 289, 410], [328, 79, 505, 266], [364, 302, 477, 432]]

def calculate_f1_score(gt_boxes, pred_boxes):
    precision = calculate_precision(gt_boxes, pred_boxes)
    recall = calculate_recall(gt_boxes, pred_boxes)
    
    # Avoid division by zero
    if precision + recall == 0:
        return 0
    
    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

f1_score = calculate_f1_score(ground_truth_boxes, predicted_boxes)

# Load an image
image_path = "data/fruit1.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (700,500))

# Create a blank image using matplotlib
fig, ax = plt.subplots(1, figsize=(8,8))
ax.set_title(f"F1 Score: {f1_score:.2f}")
plt.imshow(image)

# Draw ground truth boxes in green
for box in ground_truth_boxes:
    rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=6, edgecolor='g', facecolor='none', label='Ground Truth')
    ax.add_patch(rect)

# Draw predicted boxes in blue
for box in predicted_boxes:
    rect = patches.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1], linewidth=1, edgecolor='b', facecolor='none', label='Prediction')
    ax.add_patch(rect)

# Remove axis for better visualization
ax.axis('off')

# Show the image
plt.show()
```

### Task - use with a new image to create your own example calculating precision and recall.

```python
# List to store coordinates of the rectangles
ground_truth_boxes = []

# Initial point (where the mouse was first pressed)
rect_start = None

def draw_rectangle(action, x, y, flags, userdata):
    global rect_start
    
    # If left mouse button is pressed, record the starting point
    if action == cv2.EVENT_LBUTTONDOWN:
        rect_start = (x, y)
    
    # If left mouse button is released, record the end point and draw the rectangle
    elif action == cv2.EVENT_LBUTTONUP:
        end_point = (x, y)
        cv2.rectangle(userdata, rect_start, end_point, (255, 0, 0), 2)
        cv2.imshow("Image", userdata)
        
        # Append the coordinates in the desired format
        box = [rect_start[0], rect_start[1], x, y]
        ground_truth_boxes.append(box)
        print(f"Added box: {box}")

# Read the image and set up window
image = cv2.imread("data/fruit1.jpg")

image = cv2.resize(image, (700,500))
cv2.namedWindow("Image")
cv2.setMouseCallback("Image", draw_rectangle, image)

cv2.imshow("Image", image)
cv2.waitKey(0)

# Clean up
cv2.destroyAllWindows()

print("Selected boxes:", ground_truth_boxes)
```
