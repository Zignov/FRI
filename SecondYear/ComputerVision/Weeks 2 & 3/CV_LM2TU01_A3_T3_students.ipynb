# Task 2.1.3: Image Pre-processing for Object Detection
## Techniques for pre-processing images for object detection, including resizing, cropping, and color conversion.

The basic techniques for pre-processing images which result in better object detection further down the line.

```python
# Importing required libraries, which include:
# OpenCV library used for the many tasks in computer vision, from loading images, to processing, detecting, shapes, tracking objects etc.:
import cv2
# Numpy is very useful computation library, here we can use it to compute means:
import numpy as np
# We need the following to show images as plots (x and y axes), later we will use it to draw actual plots too:
import matplotlib.pyplot as plt
# We need the following to draw areas of interest on our plots and images:
import matplotlib.patches as patches
```

```python
# Use OpenCV to manipulate the color space of loaded image

# Load the input image
image = cv2.imread("data/lenna.png")

# Display the output image without color conversion
plt.imshow(image)
plt.title("Lenna")
plt.show()

# Apply color conversion to your image
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the output images
fig, axs = plt.subplots(1, 2, figsize=(10,10))

axs[0].imshow(imageRGB)
axs[0].set_title("Lenna RGB")

axs[1].imshow(imageGray, cmap='gray')
axs[1].set_title("Lenna Gray")
plt.show()
```

```python
# Using different tools to resize the image:
# Resizing while maintaining aspect ratio
# Scaling image
# Bilinear interpolation

# Load the image
image = cv2.imread("data/lenna.png")

# Resizing while maintaining aspect ratio
desired_width = 800
aspect_ratio = image.shape[1] / image.shape[0]
desired_height = int(desired_width / aspect_ratio)
resized_aspect = cv2.resize(image, (desired_width, desired_height))
plt.imshow(cv2.cvtColor(resized_aspect, cv2.COLOR_BGR2RGB))
plt.title('Resized (Maintaining Aspect Ratio)')
plt.show()

# Scaling: Changing the dimensions of an image by a specific scaling factor
scaling_factor = 0.1  # Scaling factor of 0.1
resized_scaled = cv2.resize(image, None, fx=scaling_factor, fy=scaling_factor)
plt.imshow(cv2.cvtColor(resized_scaled, cv2.COLOR_BGR2RGB))
plt.title('Resized (Scaling)')
plt.show()

# Resizing with bilinear interpolation
new_width = 1200
new_height = 800
resized_interpolated = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(resized_interpolated, cv2.COLOR_BGR2RGB))
plt.title('Resized (Bilinear Interpolation)')
plt.show()
```

```python
# Cropping images in Python

# Load the image and display original image for reference
image = cv2.imread("data/lenna.png")

plt.plot()
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

# Define the amount to crop from each edge (in pixels)
crop_amount = 50

# Perform edge cropping
cropped_image_edge = image[crop_amount:-crop_amount, crop_amount:-crop_amount]

# Display the cropped edges image
plt.plot()
plt.imshow(cv2.cvtColor(cropped_image_edge, cv2.COLOR_BGR2RGB))
plt.title('Cropped Edges')
plt.show()

# Define the region of interest (ROI) coordinates and dimensions
x = 100
y = 200
width = 300
height = 200

# Perform cropping
cropped_image_roi = image[y:y+height, x:x+width]

# Display the cropped ROI image
plt.plot()
plt.imshow(cv2.cvtColor(cropped_image_roi, cv2.COLOR_BGR2RGB))
plt.title('Cropped ROI')

plt.show()
```

Task 1 - Pick an image of your choosing and apply the techniques of image pre-processing to achieve the following results:

1. Resize your image to two different sizes:
    a. 800 x 600 pixels

2. Cropping:
    a. Crop out the center of the image to create a square with side length 200 pixels.

3. Color Conversion:
    a. Convert the given image to grayscale.
    b. Convert the grayscale image back to RGB where the original color information is lost and the entire image is composed of gray shades only.
    c. Display the grayscale image and the back-to-RGB image side by side.

```python
# Define a function to resize the image
def resize_image(img, width, height):
    img = cv2.resize(img, (width, height), interpolation=cv2.INTER_LINEAR)
    plt.imshow(img)
    plt.show()

# Define a function to crop the image
def crop_center(img, side_length):
    pol = int(side_length/2)
    #img = img[int(img.shape[0]/2-pol:height), int(img.shape[1]/2-pol:height)]
    #print(img.shape)
    w = int(img.shape[0]/2 - pol)
    h = int(img.shape[1]/2 - pol)
    
    
    img = img[w:w+pol, h:h+pol]
    
    plt.imshow(img)
    plt.show()

# Define a function to convert image to grayscale
def convert_to_grayscale(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #plt.imshow(img, cmap="gray")
    #plt.show()
    return img

# Define a function to convert image to RGB
def grayscale_to_rgb(img_gray):
    img_gray = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
    return img_gray

# Load an image
#image = cv2.imread("data/sheep2.png")
image = cv2.imread("data/sheep2.jpg")


# Resizing
resize_image(image, 800, 600)
resize_image(image, 400, 300)
# Cropping
crop_center(image, 200)
# Color Conversion
convert_to_grayscale(image)
# Display images using plt

figure, axes = plt.subplots(1,2)
gray = convert_to_grayscale(image)
axes[0].imshow(gray, cmap = "gray")
axes[0].set_title("gray")

rgb = grayscale_to_rgb(gray)
axes[1].imshow(rgb)
axes[1].set_title("rgb")

```
