```python
# TASK 1: Minimum bounding rectangles

# Modify the given code to draw the bounding rectangles in color on the original image, rather than on the grayscale image. Experiment with different colors for the rectangles, and display both the original image and the image with bounding rectangles side by side for comparison.

# Hint: You will need to use the cv2.drawContours function on the original img variable rather than img_gray. Experiment with different color values in the (255, 255, 255) tuple.


# Load the image
img = cv2.imread('data/trucks1.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a threshold to obtain a binary image 
# Change the threshold values in the line below to see the different effects
ret, thresh = cv2.threshold(img_gray, 170, 255, cv2.THRESH_BINARY)

plt.figure()
plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.show()

# Find contours in the binary image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


fig, sliki = plt.subplots(1,2)

sliki[0].imshow(img)
sliki[0].set_title("original")
sliki[0].axis("off")

# Iterate over each contour
for contour in contours:
    # Find the minimum bounding rectangle
    rect = cv2.minAreaRect(contour)
    
    # Convert the rectangle coordinates to integers
    box = np.intp(cv2.boxPoints(rect))
    
    # Draw the minimum bounding rectangle on the image
    cv2.drawContours(img, [box], 0, (255, 0, 255), 5) # img

# Display the image with minimum bounding rectangles
#plt.figure()
#plt.imshow(img_gray) # plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))



sliki[1].imshow(img)
sliki[1].set_title("Boxes")
sliki[1].axis("off")

plt.tight_layout()
plt.show()

```
