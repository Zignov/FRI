# Task 3.1.3: Object Tracking VS Object Detection
### Comparing Object Tracking and Object Detection methods

```python
# Importing required libraries, which include:
# OpenCV library used for the many tasks in computer vision, from loading images, to processing, detecting, shapes, tracking objects etc.:
import cv2
```

```python
# TASK1: "Tracking" with detection algorithms only.
# Implement a code that uses object detection Haar algorithm for detecting the face in every frame.
# Use this to detect the face in each frame in the video taken with the webcam.
# Display the video with drawn bounding box around detected faces.

# Hint: For loading the pre-trained Haar Cascade classifier for face detection you can use cv2.CascadeClassifier 
#       with filename parameter equal to cv2.data.haarcascades + "haarcascade_frontalface_default.xml"


# Load Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    # read frame from webcam
    ret, frame = cap.read()

    if not ret:
        break

    # convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5
    )

    # draw bounding boxes around detected faces
    for x, y, w, h in faces:
        cv2.rectangle(
            frame,
            (x, y),
            (x + w, y + h),
            (0, 255, 0),
            2
        )
    

    # display video
    cv2.imshow("Face detection", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# release resources
cap.release()
cv2.destroyAllWindows()

for _ in range(5):
    cv2.waitKey(1)
```

```python
# TASK2: Object tracking algorithm.
# Implement a code that detects a face in the first frame (or repeats detections until the any face is detected). 
# For face detection use object detection Haar algorithm (see previous snippet).
# Initialize tracker KCF with firstly detected face and continue to track the face.
# Real-time video should be captured with the webcam.
# Display the video with drawn bounding box around tracked face.


# Load Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Open webcam
cap = cv2.VideoCapture(0)

# Create KCF tracker
if hasattr(cv2, "legacy"):
    tracker = cv2.legacy.TrackerKCF_create()
else:
    tracker = cv2.TrackerKCF_create()

tracker_initialized = False

while True:
    ret, frame = cap.read()

    if not ret:
        break

    if not tracker_initialized:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5
        )

        if len(faces) > 0:
            x, y, w, h = faces[0]

            bbox = (x, y, w, h)
            tracker.init(frame, bbox)
            tracker_initialized = True

            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    else:
        success, bbox = tracker.update(frame)

        if success:
            x, y, w, h = [int(i) for i in bbox]
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        else:
            tracker_initialized = False

            if hasattr(cv2, "legacy"):
                tracker = cv2.legacy.TrackerKCF_create()
            else:
                tracker = cv2.TrackerKCF_create()

    cv2.imshow("Face tracking with KCF", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()

for _ in range(5):
    cv2.waitKey(1)
```
