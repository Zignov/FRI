```python
# TASK 2: Display video sequence with drawn predefined bounding box in each frame.

# define path for file containing predefined bounding boxes
bounding_boxes_path = "./data/got10k/groundtruth.txt"

# open the ground truth file in read mode
with open(bounding_boxes_path, mode="r", encoding="utf-8") as file:
    # read the content of the file
    data = file.read()
    # define a pattern to split the data into separate values
    split_pattern = ",|\n"
    # split the data into separate values using the pattern
    bounding_boxes = re.split(split_pattern, data)
    # convert the values to integers and store them in a numpy array
    bounding_boxes = np.array([int(float(elem)) for elem in bounding_boxes if elem != ""])
    # reshape the numpy array into a 2D array, where each row represents a bounding box
    bounding_boxes = bounding_boxes.reshape(-1, 4)

# create an empty dictionary to store file names and their corresponding bounding boxes
filenames_bboxes = {}

# initialize the index for bounding box list to 0
idx = 0

# walk through the directory of image files and their respective bounding boxes
# note: The "." means the current directory
for root, _, files in os.walk("../data/got10k/sequence_example"):
    # iterate through each file in the directory in ascending order
    for filename in sorted(files):
            # get the bounding box limits for the current image
            bounding_box_limits = bounding_boxes[idx]
            # Calculate the x and y coordinates of the top-left (x1, y1) and bottom-right (x2, y2) corners of the bounding box
            x1, y1 = bounding_box_limits[0], bounding_box_limits[1]
            x2, y2 = bounding_box_limits[0] + bounding_box_limits[2], bounding_box_limits[1] + bounding_box_limits[3]

            # Store the file name and its bounding box coordinates in the dictionary
            filenames_bboxes[filename] = [x1, y1, x2, y2]
            
            idx += 1
```

```python
# Modify the code, so that you display video sequence with drawn predefined bounding boxes around an object throughout a video sequence.
# Hint: Follow the next steps:
# - extract the top left point and bottom right point of the bounding box in each frame.
# - use function cv2.rectangle to draw the bounding box in each frame.
# - display each frame.

# set the path to the directory containing the frames
frame_dir = "./data/got10k/sequence_example/"

# get a list of all the frame filenames in the directory and sort them
frame_filenames = sorted([
    f for f in os.listdir(frame_dir)
    if f.endswith(".jpg") or f.endswith(".png")
])

# initialize index for the bounding box list to 0
bounding_boxes = []

groundtruth_path = "./data/got10k/groundtruth.txt"

with open(groundtruth_path, "r") as f:
    for line in f:
        bounding_boxes.append(list(map(float, line.strip().split(","))))

idx = 0
cv2.namedWindow("Video sequence with Bounding Boxes", cv2.WINDOW_NORMAL)

# loop through the frames
for filename in frame_filenames:
    # read the frame from file
    frame_path = os.path.join(frame_dir, filename)
    frame = cv2.imread(frame_path)

    # get the corresponding bounding box
    x, y, w, h = bounding_boxes[idx]

    # define the top left and bottom right points for the bounding box rectangle
    top_left = (int(x), int(y))
    bottom_right = (int(x+w), int(y+h))

    # draw a bounding box around the object in the image
    cv2.rectangle(frame, top_left, bottom_right, (0, 255, 0), 2)
    
    # display current frame
    cv2.imshow("Video sequence with Bounding Boxes", frame)
    
    if cv2.waitKey(30) & 0xFF == ord("q"):
        break

    # increase the frame index
    idx = idx + 1

# close the window with the image
cv2.destroyAllWindows()

for _ in range(5):
    cv2.waitKey(1)
```
