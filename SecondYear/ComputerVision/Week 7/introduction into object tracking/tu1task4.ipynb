```python
# Importing required libraries, which include:
# OpenCV library used for the many tasks in computer vision, from loading images, to processing, detecting, shapes, tracking objects etc.:
import cv2
import os
```

#### MOG2 background subtraction algorithm

```python
# TASK1: Write a program that shows background/foreground mask, using MOG2 background subtraction algorithm, in each frame of the video.

# Hint: For creating an object for MOG2 background subtraction (BackgroundSubtractorMOG2), use cv2.createBackgroundSubtractorMOG2().
#       For creating the mask use BackgroundSubtractorMOG2's apply function. 

# Set up the background subtractor object
fgbg = None

# Set the path to the directory containing the frames
frame_dir = "./data/got10k/sequence_example2/"

# Get a list of all the frame filenames in the directory and sort them
frame_filenames = sorted(os.listdir(frame_dir))

# Loop through the frames and display foreground mask
for frame in frame_filenames:
    # Read the current frame
    img = cv2.imread(os.path.join(frame_dir, frame))

    # Initialize the background subtractor if it hasn't been created yet
    if fgbg is None:
        fgbg = cv2.createBackgroundSubtractorMOG2()

    # Apply the background subtractor to get the foreground mask
    fgmask = fgbg.apply(img)

    # Display the original frame and the foreground mask
    cv2.imshow('Frame', img)
    cv2.imshow('Foreground Mask', fgmask)

    # Wait for a key press and exit if 'q' is pressed
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break
    
    
cv2.destroyAllWindows()

for _ in range(5):
    cv2.waitKey(1)
```
