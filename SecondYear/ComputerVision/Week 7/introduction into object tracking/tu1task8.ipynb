# Task 3.1.8: Image Processing Techniques in Object Tracking

```python
# Importing required libraries, which include:
# OpenCV library used for the many tasks in computer vision, from loading images, to processing, detecting, shapes, tracking objects etc.:
import cv2
# Numpy is very useful computation library:
import numpy as np
# Import pyplot from Matplotlib
# Matplotlib is a library for creating visualizations in Python.
from matplotlib import pyplot as plt
# Os is a library that provides functions for creating and removing a directory (folder), fetching its contents etc.:
import os
```

```python
# IMAGE PROCESSING TASKS

# Read the image file using OpenCV's imread() function
img = cv2.imread("./data/images/dog.jpg")
# Convert the image from BGR to RGB color space
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# TASK3: Perform image resizing. Experiment with new dimensions.

# Hint: use function cv2.resize.

# Define the new dimensions of the image
new_width = None
new_height = None

# Resize the image to the new dimensions
resized_img = np.zeros((10, 10))

# Display the original and resized images side by side
fig, axs = plt.subplots(1, 2)  # Create a figure with one row and two columns
axs[0].imshow(img)
axs[0].set_title("Original Image")
axs[1].imshow(resized_img)
axs[1].set_title("Resized Image")
# Add a title to the figure
fig.suptitle("Comparison of Original and Resized Image")
# Show the figure
plt.show()
```

```python
# TASK4: Perform image cropping. Experiment with different coordinates of the Region of Interest (ROI).

# Hint: cv2.imread function reads and returns the image as a Numpy array, 
#       so you can use numpy slicing to extract the Region of Interest (ROI) from the image.

# Define the coordinates of the top left point of the ROI
top_left_x = 500
top_left_y = 1600
# Define the coordinates of the bottom right point of the ROI
bottom_right_x = 3800
bottom_right_y = 6800

# Extract the ROI from the image
cropped_img = img[top_left_y:bottom_right_y, top_left_x:bottom_right_x]

# Display the original and cropped images side by side
fig, axs = plt.subplots(1, 2)  # Create a figure with one row and two columns
axs[0].imshow(img)
axs[0].set_title("Original Image")
axs[1].imshow(cropped_img)
axs[1].set_title("Cropped Image")
# add a title to the figure
fig.suptitle("Comparison of Original and Cropped Image")
# show the figure
plt.show()
```

```python
# TASK5: Perform color conversion. Convert the color space of the image from RGB to grayscale.

# Hint: use cv2.cvtColor function.

# Convert the image to grayscale.
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Display the original and grayscale images side by side
fig, axs = plt.subplots(1, 2)  # Create a figure with one row and two columns
axs[0].imshow(img)
axs[0].set_title("RGB Image")
axs[1].imshow(img_gray, cmap="gray")
axs[1].set_title("Grayscale Image")
# add a title to the figure
fig.suptitle("Comparison of RGB and Grayscale Image")
# show the figure
plt.show()
```

```python
# TASK6: Perform image normalization. Normalize image values to a specific range, experiment with the range intervals.
#        Try [0, 1] normalization, as it is crucial in computer vision.

# Hint: learn of min-max normalization and implement it yourself.
normalized_img = img_gray.astype(np.float32)

min = np.min(normalized_img)
max = np.max(normalized_img)

normalized_img = (normalized_img - min) / (max - min)

# Display the original and normalized images side by side
fig, axs = plt.subplots(1, 2)  # Create a figure with one row and two columns
axs[0].imshow(img)
axs[0].set_title("RGB Image")
axs[1].imshow(normalized_img, cmap="gray")
axs[1].set_title("Normalized Image")
# add a title to the figure
fig.suptitle("Comparison of RGB and Normalized Image")
# show the figure
plt.show()
```
