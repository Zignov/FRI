## Loading, displaying, storing images

```python
# Importing required libraries, which include:
# OpenCV library used for the many tasks in computer vision, from loading images, to processing, detecting, shapes, tracking objects etc.:
import cv2
# Another very useful library intended for image handling and processing. Some functions correlate to OpenCV's, some are added and easier to use:
import PIL # We could also import it as from PIL import Image, and then use Image directly
# Numpy is very useful computation library, here we can use it to compute means:
import numpy as np
# The following library will enable us to get data (images) from the web:
import urllib
# We need the following to show images as plots (x and y axes), later we will use it to draw actual plots too:
import matplotlib.pyplot as plt
```

```python
"""
Load an image from disk
"""
img = cv2.imread("data/doggo.jpg")

"""
Display the image using matplotlib.pyplot aliased as plt in the beginning
"""
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
  plt.imshow(img)
  plt.show()
```

```python
# We can load it differently - using PIL library instead of CV2:
img_rgb = PIL.Image.open("data/doggo.jpg") # This is now RGB instead of BGR
plt.imshow(img_rgb)
plt.show()

# Or we can switch channels using cv2.cvtColor:
img = cv2.imread("data/doggo.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()

# DO IT YOURSELF HERE:
img = cv2.imread("data/doggo.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
"""
Convert the image to grayscale
"""
img = cv2.imread("data/doggo.jpg")
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# And display it
plt.imshow(img_grayscale, cmap="gray")
plt.show()
```

```python
img = cv2.imread("data/doggo.jpg")
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2RGBA)

plt.imshow(img_grayscale, cmap="gray")
plt.show()
```

```python
"""
Load an image directly from the internet
"""

url = "https://hips.hearstapps.com/autoweek/assets/s3fs-public/490259.jpg?resize=980:*&keepGifs=1"
url_img = urllib.request.urlopen(url)
img_internet = PIL.Image.open(url_img)

# And display it:
plt.imshow(img_internet)
plt.show()
```

```python
"""
Load an image directly from the webcam
"""
cap = cv2.VideoCapture(0)
ret, img_camera = cap.read() # This image is in CV2 format, which means BGR
cap.release()

# Display the webcam image:
plt.imshow(cv2.cvtColor(img_camera, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
"""
Store the image to disk
"""
# For PIL images:
img_internet.save("my_beautiful_internet_image.jpg")

# Or if you loaded it using CV2:
cv2.imwrite("my_beautiful_webcam_image.jpg", img_camera)
```
