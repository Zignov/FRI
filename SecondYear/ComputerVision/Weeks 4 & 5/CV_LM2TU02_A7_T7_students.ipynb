## Exploring the essential hyperparameters in YOLOv5 and optimizing them for improved model performance

```python
import torch
from PIL import Image

# Load the trained model
model = torch.hub.load('C:/Users/vasja/Desktop/AIM4VET/LM2TU2/yolo/yolov5/', 'custom', path='C:/Users/vasja/Desktop/AIM4VET/LM2TU2/yolo/yolov5/runs/train/my_first_model3/weights/best.pt', source='local')

# Load a test image
img = Image.open("C:/Users/vasja/Desktop/AIM4VET/LM2TU2/cardata/my_testing/vid_4_600.jpg")

# Perform inference
results = model(img)

# Display results
results.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Convert results to numpy array
result_np = results.xyxy[0].cpu().numpy()

# Create a new matplotlib figure and axes
fig, ax = plt.subplots(1)

# Display the image
ax.imshow(img)

# Loop through each prediction
for det in result_np:
    x1, y1, x2, y2, conf, cls = det
    width = x2 - x1
    height = y2 - y1
    
    # Create a rectangular patch representing the bounding box and add it to the axes
    rect = patches.Rectangle((x1, y1), width, height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    
    # Add the class label and confidence score to the bounding box
    ax.text(x1, y1, f'{model.names[int(cls)]} {conf:.2f}', color='white', bbox=dict(facecolor='red', alpha=0.5))

# Display the figure
plt.show()
```
